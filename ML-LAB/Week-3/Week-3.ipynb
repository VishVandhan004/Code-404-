```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
ds = pd.read_csv('covid_worldwide.csv')
```

```python
df = pd.DataFrame(ds)
X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
print(X)
print(y)
```

```python
df.isnull().sum() #returns the missing data in the dataset
```

```python
df1 = df.copy()
print("Before:",df1.shape)

# dropping rows with missing values
df1.dropna(inplace=True)

#summarize the shape of data with missing rows removed..
print("After:",df1.shape)
```
